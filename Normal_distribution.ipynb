```python
 import numpy as np
 import matplotlib.pyplot as plt
 from scipy.stats import norm
```

```python
 x = np.linspace(-4, 4, 1000)
 mean = 0
 std_dev = 1
 
 y = norm.pdf(x, loc=mean, scale=std_dev)

 plt.plot(x, y)
 plt.xlabel("x")
 plt.title("Standard Normal Distribution (μ=0, σ=1)")
 plt.ylabel("Probability Density")
 plt.grid(True)
 plt.show()
```
